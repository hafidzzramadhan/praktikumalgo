```python
def baca_data():
    with open('NILAI.txt', 'r') as file:
        print(file.read())

def hitung_rata_rata(nama):
    with open('NILAI.txt', 'r') as file:
        lines = file.readlines()
        total_nilai = 0
        jumlah_data = 0
        for line in lines:
            data = line.strip().split()
            if len(data) >= 4 and data[0].lower() == nama.lower():
                nilai = [float(x) for x in data[1:]]
                rata_rata = sum(nilai) / len(nilai)
                print(f"Rata-rata nilai praktikum untuk {nama}: {rata_rata:.2f}")
                return
        print(f"Tidak ada data untuk {nama}.")

def update_nilai_mahasiswa(nama):
    with open('NILAI.txt', 'r') as file:
        lines = file.readlines()
    with open('NILAI.txt', 'w') as file:
        for line in lines:
            data = line.strip().split()
            if len(data) >= 4 and data[0].lower() == nama.lower():
                try:
                    nilai_praktikum_baru = input(f"Masukkan nilai praktikum baru untuk {nama} (pisahkan dengan spasi): ")
                    nilai_praktikum_baru = [float(x) for x in nilai_praktikum_baru.split()]
                    if len(nilai_praktikum_baru) == 3:
                        line = f"{nama} - {nilai_praktikum_baru[0]} {nilai_praktikum_baru[1]} {nilai_praktikum_baru[2]}\n"
                    else:
                        print("Masukkan tiga nilai praktikum.")
                        continue
                except ValueError:
                    print("Masukkan angka yang valid.")
                    continue
            file.write(line)
    print(f"Data nilai untuk {nama} berhasil diperbarui.")

def simpan_perubahan():
    print("Perubahan berhasil disimpan.")

def main():
    while True:
        print('''
        [MENU]
        1. Baca Data
        2. Mencari Nilai Rata-rata Praktikum Mahasiswa
        3. Update Nilai Mahasiswa
        4. Simpan Perubahan
        5. Exit
        ''')
        try:
            choose_menu = int(input('Silahkan Pilih: '))
        except ValueError:
            print("Masukkan angka yang valid.")
            continue

        if choose_menu == 1:
            baca_data()
        elif choose_menu == 2:
            nama = input("Masukkan nama mahasiswa: ")
            hitung_rata_rata(nama)
        elif choose_menu == 3:
            nama = input("Masukkan nama mahasiswa: ")
            update_nilai_mahasiswa(nama)
        elif choose_menu == 4:
            simpan_perubahan()
        elif choose_menu == 5:
            break
        else:
            print("Pilihan tidak valid. Silahkan coba lagi.")

if __name__ == '__main__':
    main()
```
